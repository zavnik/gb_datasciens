Импортируйте библиотеки pandas и numpy.
Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. Создайте датафреймы X и Y из этих данных.
Разбейте эти датафреймы н а тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки
составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.
Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.
Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.
Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
boston = load_boston()
```

```python
print(boston.keys())
print(boston.feature_names)
pd.DataFrame(boston.data).head()
```

```python
X = pd.DataFrame(boston['data'], columns = boston['feature_names'])
X.head()
```

```python
Y = pd.DataFrame(boston['target'], columns = ['price'])
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
X_test.head()
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
Y_predict = lr.predict(X_test)
print(Y_predict)
```

```python
result = pd.DataFrame({
    'Y_test' : Y_test['price'],
    'Y_pred' : Y_predict.flatten(),
    'error' : Y_test['price'] - Y_predict.flatten()
})
result.head()
```

```python
r2_score_result = r2_score(result['Y_test'], result['Y_pred'])
print(r2_score_result)
```
